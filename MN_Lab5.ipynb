# Metoda trapezów

```python
def trapez(f,a,b,n):
  c = 0
  h = (b-a)/n
  for i in range(n-1):
    x = a + i*h
    y = a + (i+1)*h
    c = c + (f(x)+f(y))
  return c*h/2  
```

# Metoda Simpsona

```python
def simpson(f,a,b,m):
  c = 0
  c1 = 0
  c2 = 0
  n = 2*m
  h = (b-a)/n
  for i in range(m-1):
    c1 = c1 + f(a + (2*i+1)*h)
  for i in range(1,m-1):
    c2 = c2 + f(a + (2*i)*h)
  c = h/3 * (f(a) + 4*c1 + 2*c2 + f(b))
  return c
```
