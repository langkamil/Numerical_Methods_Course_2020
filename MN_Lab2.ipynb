```python
import math
import numpy as np 
import pandas as pd
import sympy as sym #biblioteka potrzebna do pochodnej
x = sym.Symbol('x') #ustawiony symbol pochodnej
```

# ***Metoda bijekcji***

```python

#wartosci poczatkowe algorytmu z parametrami
M = 30
delta = 10**(-6)
epsilon = 10**(-24)
a1 = -1 
b1 = 4
a2 = -3
b2 = 2
a3 = -1
b3 = 5

bisection_df = pd.DataFrame(index=range(0, M+1), columns=['M','c_1','f1(x)','c_2','f2(x)','c_3','f3(x)'])
for k in range(M+1):
  bisection_df['M'][k]=k 

def function1(x): 
  f1=sym.exp(x)-1.5-sym.atan(x) #wprowadzamy funkcję dla której chcemy wyliczyć zera
  return f1  



u = function1(a1)
v = function1(b1)
e = b1 - a1

#what if function never reaches point 0
if (np.sign(u) == np.sign(v)):
  print("Wrong range!")
else:
  for i in range(M+1):
    e = e/2
    c = a1 + e
    w = function1(c)
    bisection_df['f1(x)'][i] = w
    bisection_df['c_1'][i] = c

#if we are very close to zero we end the whole procedure
    if ((abs(e) < delta) or (abs(w) < epsilon)):
      for u in range(i, M+1):
        bisection_df['f1(x)'][u] = 0
        bisection_df['c_1'][u] = c
      break
#what if not          
    if (np.sign(u) != np.sign(v)):
      b1 = c
      v = w
    else:
      a1 = c
      u = w


def function2(x): 
  f2=x**3 + 94*x**2 - 389*x +294 #wprowadzamy funkcję dla której chcemy wyliczyć zera
  return f2 

u = function2(a2)
v = function2(b2)
e = b2 - a2

#what if function never reaches point 0
if (np.sign(u) == np.sign(v)):
  print("Wrong range!")
else:
  for i in range(M+1):
    e = e/2
    c = a2 + e
    w = function1(c)
    bisection_df['f2(x)'][i] = w
    bisection_df['c_2'][i] = c

#if we are very close to zero we end the whole procedure
    if ((abs(e) < delta) or (abs(w) < epsilon)):
      for u in range(i, M+1):
        bisection_df['f2(x)'][u] = 0
        bisection_df['c_2'][u] = c
      break
#what if not          
    if (np.sign(u) != np.sign(v)):
      b2 = c
      v = w
    else:
      a2 = c
      u = w

def function3(x): 
  f3=x**3 - 5*x**2 + 3*x -7 #wprowadzamy funkcję dla której chcemy wyliczyć zera
  return f3      

u = function3(a3)
v = function3(b3)
e = b3 - a3

#what if function never reaches point 0
if (np.sign(u) == np.sign(v)):
  print("Wrong range!")
else:
  for i in range(M+1):
    e = e/2
    c = a3 + e
    w = function1(c)
    bisection_df['f3(x)'][i] = w
    bisection_df['c_3'][i] = c

#if we are very close to zero we end the whole procedure
    if ((abs(e) < delta) or (abs(w) < epsilon)):
      for u in range(i, M+1):
        bisection_df['f3(x)'][u] = 0
        bisection_df['c_3'][u] = c
      break
#what if not          
    if (np.sign(u) != np.sign(v)):
      b3 = c
      v = w
    else:
      a3 = c
      u = w      



bisection_df
```

#  ***Metoda Newtona***

**> Szczegółowy algorytm Newtona z parametrami delta i epsilon**



```python
#wartosci poczatkowe algorytmu z parametrami
M = 15
delta = 10**(-6)
epsilon = 10**(-24)
```



**> Określenie tablicy do zapisu wyników**


```python
newton_df = pd.DataFrame(index=range(0,M+1), columns=['M','w_p1','f1(x)','w_p2','f2(x)','w_p3','f3(x)'])
for k in range(M+1):
    newton_df['M'][k]=k 
```

* zdefiniowanie funkcji oraz jej pochodnej
1. f1(x) = e^x - 1.5 - arctg(x) dla w_p = -7
2. f2(x) = x^3 + 94x^2 - 389x +294 dla w_p = 2
3. f3(x) = x^3 - 5x^2 + 3x -7 dla w_p = 5 


```python
def function1(x): 
    f1=sym.exp(x)-1.5-sym.atan(x) #wprowadzamy funkcję dla której chcemy wyliczyć zera
    return f1  
dx1 = sym.diff(function1(x), x) #pochodna tej funkcji
ddx1 = sym.lambdify(x, dx1) #obliczanie pochodnej w punkcie x

w_p1 = -7.0

def function2(x): 
    f2=x**3 + 94*x**2 - 389*x +294 #wprowadzamy funkcję dla której chcemy wyliczyć zera
    return f2  
dx2 = sym.diff(function2(x), x) #pochodna tej funkcji
ddx2 = sym.lambdify(x, dx2) #obliczanie pochodnej w punkcie x

w_p2 = 4.0

def function3(x): 
    f3=x**3 - 5*x**2 + 3*x -7 #wprowadzamy funkcję dla której chcemy wyliczyć zera
    return f3  
dx3 = sym.diff(function3(x), x) #pochodna tej funkcji
ddx3 = sym.lambdify(x, dx3) #obliczanie pochodnej w punkcie x

w_p3 = 5.0
```

```python
#algorytm

w_p1 = float(w_p1)
f1 = function1(w_p1)
for q in range(M+1):
  newton_df['f1(x)'][q]=f1
  newton_df['w_p1'][q]=w_p1
  if ddx1(w_p1)==0:
    break
  w_d1 = float(w_p1 - f1/ddx1(w_p1))
  f1 = function1(w_d1) 
  if (abs(w_d1-w_p1) < delta or abs(f1) < epsilon):
    for r in range(q, M+1):
      newton_df['f1(x)'][r]=0
      newton_df['w_p1'][r]=w_p1
    break
  w_p1 = w_d1
  
  

w_p2 = float(w_p2)
f2 = function2(w_p2)
for q in range(M+1):
  newton_df['f2(x)'][q]=f2
  newton_df['w_p2'][q]=w_p2
  if ddx2(w_p2)==0:
    break
  w_d2 = float(w_p2 - f2/ddx2(w_p2))
  f2 = function2(w_d2)
  if (abs(w_d2-w_p2) < delta or abs(f2) < epsilon):
    for r in range(q, M+1):
      newton_df['f2(x)'][r]=0
      newton_df['w_p2'][r]=w_p2
    break
  w_p2 = w_d2

w_p3 = float(w_p3)
f3 = function3(w_p3)
for q in range(M+1):
  newton_df['f3(x)'][q]=f3
  newton_df['w_p3'][q]=w_p3
  if ddx3(w_p3)==0:
    break
  w_d3 = float(w_p3 - f3/ddx1(w_p3))
  f3 = function3(w_d3)
  if (abs(w_d3-w_p3) < delta or abs(f3) < epsilon):
    for r in range(q, M+1):
      newton_df['f3(x)'][r]=0
      newton_df['w_p3'][r]=w_p3
    break
  w_p3 = w_d3 


  #wyniki
newton_df
```

# ***Metoda siecznych***

```python
M=15
delta=10**(-6)
epsilon=10**(-24)
```

```python
sieczne_df = pd.DataFrame(index=range(0,M+1), columns=['M','a1','f1(a1)','a2','f2(a2)','a3','f3(a3)'])
for k in range(M+2):
    sieczne_df['M'][k]=k
```

```python
def function1(x): 
    f1=sym.exp(x)-1.5-sym.atan(x) 
    return f1  
a1=-20
b1=-5
def function2(x): 
    f2=x**3 + 94*x**2 - 389*x +294 
    return f2  
a2=-5
b2=10

def function3(x): 
    f3=x**3 - 5*x**2 + 3*x -7 
    return f3  
a3=0
b3=10
```

```python
a1=float(a1)
b1=float(b1)
y11=function1(a1)
y21=function1(b1)
for t in range(M+2):
    sieczne_df['f1(a1)'][t]=y11
    sieczne_df['a1'][t]=a1
    s=(b1-a1)/(y21-y11)
    b1=a1
    y21=y11
    a1=a1-y11*s
    y11=function1(a1)
    if abs(y11)<epsilon or abs(b1-a1)<delta:
         for i in range(t,M+2):
           sieczne_df['f1(a1)'][i]=0
           sieczne_df['a1'][i]=a1
         break


a2=float(a2)
b2=float(b2)
y12=function2(a2)
y22=function2(b2)
for t in range(M+2):
    sieczne_df['f2(a2)'][t]=y12
    sieczne_df['a2'][t]=a2
    s=(b2-a2)/(y22-y12)
    b2=a2
    y22=y12
    a2=a2-y12*s
    y12=function2(a2)
    if abs(y12)<epsilon or abs(b2-a2)<delta:
         for i in range(t,M+2):
           sieczne_df['f2(a2)'][i]=0
           sieczne_df['a2'][i]=a2
         break


a3=float(a3)
b3=float(b3)
y13=function3(a3)
y23=function3(b3)
for t in range(M+2):
    sieczne_df['f3(a3)'][t]=y13
    sieczne_df['a3'][t]=a3
    s=(b3-a3)/(y23-y13)
    b3=a3
    y23=y13
    a3=a3-y13*s
    y13=function3(a3)
    if abs(y13)<epsilon or abs(b3-a3)<delta:
         for i in range(t,M+2):
           sieczne_df['f3(a3)'][i]=0
           sieczne_df['a3'][i]=a3
         break

sieczne_df
```

```python
import math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym #biblioteka potrzebna do pochodnej
#from google.colab import files
x = sym.Symbol('x') #ustawiony symbol pochodnej
```

**Wartości początkowe**

```python
#Przyjmujemy nastepujace wartosci poczatkowe
epsilon=10**(-24) 
delta=10**(-6)
#liczba kroków
M=30
```

**Przykłady funkcji:**
* f(x) = x^3 - 5x^2 + 3x - 7

```python
def function(x): 
    f=x**3 - 5*x**2 + 3*x - 7 #wprowadzamy funkcję dla której chcemy wyliczyć zera
    return f  
dx = sym.diff(function(x), x) #pochodna tej funkcji
ddx = sym.lambdify(x, dx) #obliczanie pochodnej w punkcie x

#poczatek przedzialu a
a=2.5
#koniec przedzialu b
b=10
#wartosc poczatkowa w_p
w_p=5
```

```python
#Stworzenie tablicy do rankingu zbieżności
df1 = pd.DataFrame(index=range(0,M+1), columns=['M','x-bisekcja','f(x)-bisekcja','x-Newton','f(x)-Newton','x-sieczne','f(x)-sieczne'])
for k in range(M+1):
    df1['M'][k]=k 

#ALGORYTM BIJEKCJA
a_bi = float(a)
b_bi = float(b)   
u = function(a_bi)
v = function(b_bi)
e = b_bi - a_bi
if (np.sign(u) == np.sign(v)):
        print("Wrong range!")
else: 
  for p in range(M+1):
    e = e/2
    c = a_bi + e
    w = function(c)
    df1['f(x)-bisekcja'][p] = w
    df1['x-bisekcja'][p] = c
    if ((abs(e) < delta) or (abs(w) < epsilon)):
      for l in range(p,M+1):
        df1['f(x)-bisekcja'][l]=0
        df1['x-bisekcja'][l]=c
      break
    if (np.sign(u) != np.sign(v)):
      b_bi = c
      v = w
    else:
      a_bi = c
      u = w

#ALGORYTM NEWTON   

w_p=float(w_p)
funkcja_newton = function(w_p)
for q in range(M+1):
  df1['f(x)-Newton'][q]=funkcja_newton
  df1['x-Newton'][q]=w_p
  if ddx(w_p)==0:
    break
  w_d = float(w_p - funkcja_newton/ddx(w_p))
  funkcja_newton = function(w_d)
  if abs(w_d-w_p) < delta or abs(funkcja_newton) < epsilon:
    for r in range(q, M+1):
      df1['f(x)-Newton'][r]=0
      df1['x-Newton'][r]=w_p
    break
  w_p = float(w_d) 

#ALGORYTM SIECZNYCH
y1=function(a)
y2=function(b)
for t in range(M+1):
    df1['f(x)-sieczne'][t]=y1
    df1['x-sieczne'][t]=a
    s=(b-a)/(y2-y1)
    b=a
    y2=y1
    a=a-y1*s
    y1=function(a)
    if abs(y1)<epsilon or abs(b-a)<delta:
         for i in range(t,M+1):
           df1['f(x)-sieczne'][i]=0
           df1['x-sieczne'][i]=a
         break

#wyniki
df1
```

```python
x_values = np.arange(-4,7,0.001)
y_values = [x**3 - 5*x**2 + 3*x - 7 for x in x_values]

plt.figure()
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 1)

for k in range(30):
    if k == 29:
        ax.scatter(df1['x-bisekcja'][k], 0, s = 30, c ='red', label='x-bisekcja')
        ax.scatter(df1['x-Newton'][k], 0, s = 30, c = 'green', label='x-Newton')
        ax.scatter(df1['x-sieczne'][k], 0, s = 30, c = 'yellow', label='x-sieczne')
    else:
        ax.scatter(df1['x-bisekcja'][k], 0, s = 30, c ='red') 
        ax.scatter(df1['x-Newton'][k], 0, s = 30, c = 'green')
        ax.scatter(df1['x-sieczne'][k], 0, s = 30, c = 'yellow')

for k in np.arange(-10, 10, 0.01):
  ax.scatter(k,0, s = 0.001, c = 'black')

ax.set_xlabel('x', fontsize = 14)
ax.set_ylabel('f(x)', fontsize = 14)
plt.legend()

plt.savefig("f1(x).png")
files.download("f1(x).png") 

plt.show()
```
